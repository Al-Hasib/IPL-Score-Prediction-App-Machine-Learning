```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drop_columns = ['mid', 'date', 'striker', 'non-striker']
df1 = df.drop(drop_columns, axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
categorical_columns = ['venue', 'bat_team', 'bowl_team', 'batsman', 'bowler']
numerical_columns = ['runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5']

for i in categorical_columns:
    print(f'\n\n column : {i}\n')
    print(df[i].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize the encoder
label_encoder = LabelEncoder()

for i in categorical_columns:
    print(f'\n\n column : {i}\n')
    df1[i] = label_encoder.fit_transform(df1[i])

df1.head()
```

```python
# Calculate Pearson correlation
df1.corr()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler


# Initialize the scaler
scaler_min_max = MinMaxScaler()

# Initialize the scaler
scaler_stc = StandardScaler()

# Fit and transform the data
df1[numerical_columns] = scaler_stc.fit_transform(df1[numerical_columns])
df1[categorical_columns] = scaler_min_max.fit_transform(df1[categorical_columns])

df1.head()
```

```python
X = df1.drop('total', axis=1)
y = df1['total']
X.head()
```

```python
from sklearn.model_selection import train_test_split
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=42)

# Print the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Make predictions
y_pred = linear_model.predict(X_test)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
import numpy as np
# Initialize the models
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=0.1)
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the models
ridge_model.fit(X_train, y_train)
lasso_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)

# Make predictions
y_pred_ridge = ridge_model.predict(X_test)
y_pred_lasso = lasso_model.predict(X_test)
y_pred_forest = forest_model.predict(X_test)

# Calculate evaluation metrics for Ridge Regression
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("\nRidge Regression Metrics:")
print("Mean Squared Error (MSE):", mse_ridge)
print("Root Mean Squared Error (RMSE):", rmse_ridge)
print("Mean Absolute Error (MAE):", mae_ridge)
print("R-squared (R):", r2_ridge)

# Calculate evaluation metrics for Lasso Regression
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
rmse_lasso = np.sqrt(mse_lasso)
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("\nLasso Regression Metrics:")
print("Mean Squared Error (MSE):", mse_lasso)
print("Root Mean Squared Error (RMSE):", rmse_lasso)
print("Mean Absolute Error (MAE):", mae_lasso)
print("R-squared (R):", r2_lasso)

# Calculate evaluation metrics for Random Forest Regression
mse_forest = mean_squared_error(y_test, y_pred_forest)
rmse_forest = np.sqrt(mse_forest)
mae_forest = mean_absolute_error(y_test, y_pred_forest)
r2_forest = r2_score(y_test, y_pred_forest)

print("\nRandom Forest Regression Metrics:")
print("Mean Squared Error (MSE):", mse_forest)
print("Root Mean Squared Error (RMSE):", rmse_forest)
print("Mean Absolute Error (MAE):", mae_forest)
print("R-squared (R):", r2_forest)
```
